```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_html(url):
    html = ""
    resp = requests.get(url)
    
    if resp.status_code == 200:
        html = resp.text
        
    return html
```

```python
# url = "https://namu.wiki/w/%EA%B0%95%EC%B1%99%EC%9D%B4"
url = "https://namu.wiki/w/%EB%B6%84%EB%A5%98:1%EC%84%B8%EB%8C%80%20%ED%8F%AC%EC%BC%93%EB%AA%AC"
html = get_html(url)

print(html)
```

```python
c = get_html(url)

soup = BeautifulSoup(c)

pages = []
table = soup.find(class_="wiki-category-container")
for li in table('li'):
    for a in li('a'):
        print("name : {}    link : {}".format(a.get_text(), a['href']))
        pages.append("https://namu.wiki" + a['href'])
        
```

```python
# for page in pages:
page = pages[9]
c = get_html(page)

soup = BeautifulSoup(c)

article = soup.find(class_="container-fluid wiki-article")
tables = article.find_all(class_="wiki-table")

for table in tables:
    utag = table.find_all("u")
    for u in utag:
        if u.get_text().strip() == "도감설명" or u.get_text().strip() == "도감 설명":
            for td in table('td'):
                for p in td('p'):
                    text = p.get_text().strip()
                    if len(text) > 7:
                        print(text)
    
# for td in tables[7]('td'):
#     for p in td('p'):
#         text = p.get_text().strip()
#         print(text)
```
