```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_html(url):
    html=""
    resp= requests.get(url)
    
    if resp.status_code ==200:
        html= resp.text
        
    return html
```

```python
url= "https://namu.wiki/w/%EB%B6%84%EB%A5%98:1%EC%84%B8%EB%8C%80%20%ED%8F%AC%EC%BC%93%EB%AA%AC"
html= get_html(url)

print(html)
```

```python
c= get_html(url)
soup= BeautifulSoup(c)

pages=[]
table= soup.find(class_='wiki-category-container')
for li in table('li'):
    for a in li('a'):
        print("name : {}    link : {}".format(a.get_text(), a['href']))
        pages.append("https://namu.wiki" + a['href'])
```

```python
i = 0
for page in pages:
    i += 1
    if i==58:
        continue
    else: 
        count=0
        tmp=4 #제일 기본 (이름, 도감번호, 성비, 타입)
        print("[{}]".format(i))
        print(page)
        c = get_html(page)

        soup = BeautifulSoup(c)

        article = soup.find(class_="container-fluid wiki-article")
        tables = article.find_all(class_="wiki-table")
        td_list= []

        for table in tables:
            tdTag = table.find_all("td")
            for td in tdTag:
                if td.get_text().strip() == "성비":
                    for tr in table('tr'):
                        for td in tr('td'):
                            count+=1

                        if count>tmp:
                            tmp = count
                            count=0
                            txt= td.get_text().strip()
                            td_list.append(txt)
                            print(td_list[-1])
                        else:
                            count=0
```
