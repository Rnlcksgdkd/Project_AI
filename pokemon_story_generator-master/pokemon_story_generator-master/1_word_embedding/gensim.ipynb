```python
# import module
import pandas as pd
import numpy as np
import tensorflow as tf
import collections

from konlpy.tag import Twitter
import re
import math
import random

import pandas as pd
```

```python
# data load
df=pd.read_csv("pk_data_total.csv")
```

```python
from konlpy.tag import *
hannanum = Hannanum()
kkma = Kkma()
twitter = Twitter()
```

```python
# 문장단위로 나누기
pk_data =df
desc_list = []
for i in range(pk_data.shape[0]):

    for desc in pk_data['desc'][i].split('.'):
        desc_list.append(desc)
```

```python
desc_list[0]
```

```python
# 형태소 기준으로 분리하기
data=[]
for sentence in desc_list:
    data.append(kkma.morphs(sentence))

data
```

```python
# gensim 
from gensim.models.word2vec import Word2Vec
```

```python
# 학습
model = Word2Vec(data,workers=4,size=30,min_count=5,window=2,sample=1e-3 )
```

```python
model.init_sims(replace=True)
```

```python
vocab=list(model.wv.vocab)
model[vocab][0]
```

```python
posVec=pd.DataFrame(model[vocab])
posVec
```

```python
posVec['pos']=vocab
```

```python
posVec.to_csv("posVec.csv", encoding='utf-8-sig')
```
