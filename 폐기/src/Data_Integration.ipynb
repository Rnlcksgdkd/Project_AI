> ## Data_1 
  - "https://raw.githubusercontent.com/ZIZUN/korean-malicious-comments-dataset/master/Dataset.csv"

```python
import pandas as pd

url_1 = "https://raw.githubusercontent.com/ZIZUN/korean-malicious-comments-dataset/master/Dataset.csv"

df_1 = pd.read_csv(url_1)
df_1.info()
df_1.head()
```

```python

texts = []
labels = []

for text in df_1['content	lable']:
  text , label = text.split('\t')
  texts.append(text)
  if int(label) == 0:
    labels.append(1)
  else: labels.append(0)
len(texts) , len(labels)

df_1['text'] = texts
df_1['label'] = labels
df_1.drop(['content\tlable'] , axis = 1, inplace = True)
df_1.head()
```

```python
df_1['label'].value_counts()
```

```python
# 데이터 1 클린 30개 (라벨 0 = 클린)
df_1[df_1['label'] == 0].head(30)
```

```python
# 데이터 1 악플 30개 (라벨 1 = 악플)
df_1[df_1['label'] == 1].head(30)
```

> ## Data 2
  - 'https://raw.githubusercontent.com/2runo/Curse-detection-data/master/dataset.txt'


```python
url_2 = 'https://raw.githubusercontent.com/2runo/Curse-detection-data/master/dataset.txt'

import keras

file_path = keras.utils.get_file('dataset.txt', url_2)
text = open(file_path , 'rb').read().decode(encoding = 'utf-8')

```

```python
# Data 2 : 5825개
sentences = text.split('\r\n')
sentences

text_2 = [ text[:-2] for text in sentences  if len(text) > 0 ]
label_2 = [ int(text[-1:]) for text in sentences if len(text) > 0]

len(text_2) , len(label_2)
```

```python
df_2 = pd.DataFrame( {'text' : text_2 , 'label' : label_2 })

df_2.head(30)
```

```python
# 클린 댓글 3781개
print(len(df_2[df_2['label'] == 0]))
df_2[df_2['label'] == 0].head(30)
```

```python
# 악플 2044개
print(len(df_2[df_2['label'] == 1]))
df_2[df_2['label'] == 1].head(30)
```

> ## Data 3
- 'https://raw.githubusercontent.com/kocohub/korean-hate-speech/master/labeled/train.tsv'

```python
url_3 = 'https://raw.githubusercontent.com/kocohub/korean-hate-speech/master/labeled/train.tsv'

df_3 = pd.read_csv(url_3 , sep = '\t')

df_3.info()
df_3.head()
```

```python
print("Bias")
print(df_3['bias'].value_counts())
print("")
print("Contain_Gender_Bias")
print(df_3['contain_gender_bias'].value_counts())
print("")

print("Hate")
print(df_3['hate'].value_counts())
print("")
```

```python
# 완전 클린한 데이터 3273개 / 악플  4623개

is_dirty = (df_3['bias'] != 'none') | (df_3['contain_gender_bias'] != False) |  (df_3['hate'] != 'none')
print(len(df_3[is_dirty]))
df_3[is_dirty].head(30)
```

```python
# bias 데이터 2748개
# gender / other 2개 타입

is_bias = (df_3['bias'] != 'none')
print(len(df_3[is_bias]))
df_3[is_bias].head(30)
```

```python
# contain_gender_bias 데이터 1232개
# hate / offensive 2종류
is_gender_bias = (df_3['contain_gender_bias'] != False)
print(len(df_3[is_gender_bias]))
df_3[is_gender_bias].head(30)
```

```python
# hate 데이터 4410개

is_hate = (df_3['hate'] != 'none')
print(len(df_3[is_hate]))
df_3[is_hate].head(30)
```

```python

df_3['label'] = is_dirty.astype(int)

df_3 = pd.DataFrame({'text' : df_3['comments'] , 'label' : df_3['label']})

# 클린데이터 3273개 / 악플데이터 4623개
print(df_3['label'].value_counts())
df_3.head()
```

# **데이터 전체 취합**
